## Introducción
<p style="text-align: justify;">La preservación del recurso agua está contemplado dentro de los ODS del programa de las naciones unidas (PNUD), que establece la importancia de
asegurar la disponibilidad y la gestión sostenible de fuentes de agua limpia para todos [1]. Sin embargo, en países en vía de desarrollo estás fuentes de agua suelen estar contaminadas, debido a la falta de control y regulación de las actividades antropogénicas que se realizan en zonas cercanas a los cuerpos hídricos[2].</p>

<p style="text-align: justify;">Ecuador es uno de los países latinoamericanos con mayor cantidad de agua dulce, sin embargo, el 40% de la población rural no posee acceso al agua para consumo humano. Otro de los problemas persistentes es que el 90% de las aguas residuales no poseen un correcto tratamiento. Lo que conlleva a que más del 50% de las fuentes de agua dulce no sean aptas para el consumo debido a su pésima calidad[3].</p>

<p style="text-align: justify;">Las actividades industriales que no poseen un correcto control de los desechos residuales que generan han provocado en gran cantidad el deterioro del medio ambiente, y han influido directamente en la calidad del agua, algunas de las principales malas prácticas ambientales son las descargas de aguas con altos niveles de metales pesados en ríos y quebradas, causando la degradación de la flora y la fauna y afectando a los usuarios que hacen uso de los ríos[4]. En este contexto, el objetivo de este proyecto es evaluar espacial y temporalmente el contenido de elementos potencialmente tóxicos en cuerpos de agua con fines de consumo, en el área minera de Santa Rosa, mediante la utilización de la analítica de datos.</p>

## Objetivo
<p style="text-align: justify;">Analizar la concentración de metales pesados en la cuenca del río Santa Rosa utilizando las bibliotecas de Python numpy, pandas, matplotlib y seaborn, para la identificación de tendencias y comportamientos durante la época seca y lluviosa del año 2021.</p>

### Funciones - Limpar Data


```python
# limpiar la data de espacios vacios y duplicados
def eliminar_duplicados(data, keep="First"):
    return data.drop_duplicates(inplace=True)
    
dup = eliminar_duplicados(rainyS_2021)
print(dup)


def espacios_vacios(data):
    data_limp = data.columns = data.columns.str.strip()
    return data_limp

sin_espacios = espacios_vacios(dryS_2021)
print(sin_espacios)
```

### Funciones - Agregar y agrupar


```python
# Agrupar y agregar información de los datos 
# Época seca
def agrupar_agregar(data, column, agg_colum, funciones_agregacion): 
    agrupado = data.groupby(column).agg({agg_colum: funciones_agregacion})
    return agrupado

agregar = agrupar_agregar(dryS_2021, "COD", "T (°C)",["sum", "mean", "count"]).reset_index()
print("Multiples agregaciones por código:")
agregar.head()
```

### Funciones - Información estadística

```python
def estadistica(data):
    return data.describe()

dr_estadist = estadistica(dryS_2021)
dr_estadist
```

### Funciones - Generador de gáficos

```python
def generador_graficos(data, codigo, elemento, xlabel, ylabel, title):
       
    fig, ax = plt.subplots(figsize=(14,8))

    ax.plot(dryS_2021[codigo],dryS_2021[elemento])
    
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    ax.grid()
    
    ax.fill_between(dryS_2021[codigo], dryS_2021[elemento] , color="pink", alpha=0.33)
    
    ax.minorticks_on()
    ax.xaxis.set_minor_locator(ticker.MultipleLocator(10))
    ax.yaxis.set_minor_locator(ticker.MultipleLocator(500))
    
    ax.grid(which='minor', linestyle=':', linewidth='0.5', color='pink')
    
    ax.grid()
    ax.minorticks_on()
    ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_minor_locator(ticker.MultipleLocator(5))
    ax.grid(which='minor', linestyle=':', linewidth='0.5', color='green')
    plt.xticks(rotation=45) #mejora la legibilidad de las etiquetas del eje x
    plt.tight_layout() #ajustar automáticamente los parámetros de los ejes para que el contenido de los gráficos se ajuste bien a la figura.
    
    plt.show()

graficos = generador_graficos(dryS_2021, "COD", "Al", xlabel= "Código de muestra", ylabel='Concentración de Al (mg/L)', title='Variación de la concentración de Al, a lo largo del río Santa Rosa')
graficos
```

### Conclusiones
1. Se identificó que en ambas épocas del año 2021, el elemento con mayor concentración en las muestras de agua superficial fue el calcio.
2. De forma general, las concentración de iones y metales durante la época seca fueron mayor que durante la época lluviosa.
3. La concentración de sales y metales puede distribuirse a lo largo del río Santa Rosa debio al aumento del nivel del agua en los ríos por las lluvias.
4. Se identificó que los valores máximos de sales y metales se encontraban en las muestras que fueorn tomadas en las quebradas, esto puede deberse a la presencia de actividad minera.
